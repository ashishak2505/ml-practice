```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt 
import seaborn as sn
%matplotlib inline
bank_df= pd.read_csv(r'C:\Users\Ashish\Desktop\machine learning\bank.csv')
bank_df.head(5)
```

```python
bank_df.info()
```

```python
X_features= list(bank_df.columns)
X_features.remove('subscribed')
X_features
```

```python
encoded_bank_df = pd.get_dummies(bank_df[X_features], drop_first =  True)
```

```python
boolean_columns = encoded_bank_df.select_dtypes(include= ['bool']).columns
for col in boolean_columns:
    encoded_bank_df[col] = encoded_bank_df[col].astype(int)
```

```python
Y = bank_df.subscribed.map(lambda x : int(x=="yes"))
X = encoded_bank_df
```

```python
logit_model = sm.Logit(Y, sm.add_constant(X)). fit()
```

```python
logit_model.summary2()
```

```python
def get_significant_vars(lm):
    var_p_vals_df= pd.DataFrame(lm.pvalues) # store the pvalues and corresponding  names in a dataframe
    var_p_vals_df['vars'] = var_p_vals_df.index
    var_p_vals_df.columns = ['pvals', 'vars']
    return list(var_p_vals_df[var_p_vals_df.pvals<=0.05]['vars'])
```

```python
significant_vars= get_significant_vars(logit_model)
significant_vars
```

```python
X_features = [
 'current-campaign',
 'previous-campaign',
 'job_retired',
 'marital_married',
 'education_tertiary',
 'housing-loan_yes',
 'personal-loan_yes']
```

```python
logit_model_2 = sm.Logit(Y, sm.add_constant(X[X_features])).fit()
```

```python
logit_model_2.summary2()
```

```python
Y_pred_df = pd.DataFrame({'actual' : Y, 'predicted_prob' : logit_model_2.predict(sm.add_constant(X[X_features]))})
```

```python
sorted_predicted_df = Y_pred_df[['predicted_prob' , 'actual']].sort_values('predicted_prob', ascending = False)
```

```python
num_per_decile = int(len(sorted_predicted_df)/10) 
print("Number of observation per decile: " , num_per_decile)
```

```python
def get_decile(df):
    #set first decile
    df['decile'] = 1

    idx = 0
    for each_d in range(0,10):
        df.iloc[idx: idx+num_per_decile, df.columns.get_loc('decile')] = each_d
        idx+= num_per_decile
    df['decile'] = df['decile'] +1
    return df
    
```

```python
deciles_predict_df = get_decile(sorted_predicted_df)
```

```python
deciles_predict_df[0:10]
```

```python
gain_lift_df = pd.DataFrame(deciles_predict_df.groupby('decile')['actual']. sum()).reset_index()
gain_lift_df.columns = ['decile', 'gain']
```

```python
gain_lift_df['gain_percentage'] = (100* gain_lift_df.gain.cumsum()/ gain_lift_df.gain.sum())
```

```python
gain_lift_df
```

```python
plt.figure(figsize = (8,4))
plt.plot (gain_lift_df['decile'], gain_lift_df['gain_percentage'], '-')
plt.show()
```

```python
gain_lift_df['lift'] = (gain_lift_df.gain_percentage /(gain_lift_df.decile *10))
gain_lift_df
```

```python
plt.figure (figure = (8,4))
plt.plot(gain_lift_df['decile'], gain_lift_df['lift'], '-')
plt.show()
```
