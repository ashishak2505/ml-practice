```python
X_feature = list(diabeties_df.columns)
X_feature.remove('Diabetes_Diagnosis')
X_feature
```

```python
encoded_df = pd.get_dummies(diabeties_df[X_feature], drop_first= True)
```

```python
list(encoded_df)
```

```python
boolean_columns = encoded_df.select_dtypes(include = ['bool']).columns
for col in boolean_columns:
    encoded_df[col] = encoded_df[col].astype(int)
```

```python
encoded_df.info()
```

```python
Y = diabeties_df.Diabetes_Diagnosis
X = encoded_df
```

```python
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size= 0.3, random_state= 42)
```

```python
X_scaled= X_scaler.fit_transform(X)
```

```python
print(X_scaled)
```

```python
from sklearn.svm import SVC
```

```python
clf =SVC()
clf.fit(train_X, train_Y)
```

```python
pred_Y = clf.predict(test_X)
```

```python
from sklearn import metrics
def draw_cm(actual, predicted):
    cm = metrics.confusion_matrix(actual, predicted, labels= [1,0])
    sns.heatmap(cm, annot= True, fmt = '.2f',
               xticklabels= ["Diagnosed", "Not Dignosed"],
    yticklabels = ["Dignosed", "Not Dignoseed"])
    plt.ylabel("True label")
    plt.xlabel("predicted label")
    plt.show()
```

```python
cm = draw_cm(test_Y, pred_Y)
```

```python
print(metrics.classification_report(test_Y, pred_Y))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

param_grid = {
    'C': [0.1, 1, 10],
    'gamma': [1, 0.1, 0.01],
    'kernel': ['rbf', 'linear']
}

grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=2)
grid.fit(train_X, train_Y)

print(grid.best_params_)
```
